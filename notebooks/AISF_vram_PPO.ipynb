```python
import gymnasium as gym
from stable_baselines3 import PPO
from stable_baselines3.common.callbacks import EvalCallback
from gymnasium.wrappers import RecordVideo
from pyvirtualdisplay import Display
import os

# 1. Start virtual display (Required for Colab rendering)
if 'virt_display' not in locals():
    virt_display = Display(visible=0, size=(1400, 900))
    virt_display.start()

# 2. Setup Environments
env_id = "BipedalWalker-v3"
train_env = gym.make(env_id)

# Create evaluation environment to record videos periodically
eval_env = gym.make(env_id, render_mode="rgb_array")
eval_env = RecordVideo(eval_env, video_folder="./logs/videos",
                       episode_trigger=lambda x: True, name_prefix="eval")

# 3. Setup Callbacks
# This checks the model every 25,000 steps and saves a video
eval_callback = EvalCallback(eval_env,
                             best_model_save_path="./logs/best_model",
                             log_path="./logs/results",
                             eval_freq=25000,
                             deterministic=True,
                             render=False)

# 4. Define and Train the Model
model = PPO("MlpPolicy", train_env, verbose=1)

print("Starting training. Look below for the live progress bar!")

# --- THE PROGRESS BAR IS ENABLED HERE ---
model.learn(
    total_timesteps=100000,
    callback=eval_callback,
    progress_bar=True  # This creates the visual bar in Colab
)

# 5. Save final progress
model.save("ppo_bipedal_final")
print("\nTraining complete! Use the video display script to see the walker's progress.")
```

```python
import base64
from pathlib import Path
from IPython import display as ipythondisplay

def show_latest_video(video_folder="./logs/videos"):
    video_files = sorted(Path(video_folder).glob("*.mp4"), key=os.path.getmtime)
    if video_files:
        video_path = video_files[-1] # Show the most recent one
        print(f"Displaying: {video_path}")
        video_b64 = base64.b64encode(video_path.read_bytes())
        ipythondisplay.display(ipythondisplay.HTML(data=f'''
            <video alt="test" autoplay loop controls style="height: 400px;">
                <source src="data:video/mp4;base64,{video_b64.decode('ascii')}" type="video/mp4" />
            </video>'''))
    else:
        print("No videos found yet. Keep training!")

show_latest_video()
```
